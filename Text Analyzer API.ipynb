```python
import sys
!conda install --yes --prefix {sys.prefix} PyJWT

import requests
import json
import jwt
from datetime import datetime, timedelta

def extract_text(url, access_token):
    return requests.post(
        'https://www.jstor.org/api/labs-text-extractor-service/v2',
        headers={
            'User-Agent': 'Python API client',
            'Authorization': 'JWT %s' % access_token},
        json={'url': url}
    ).json()

def analyze_text(text, access_token):
    return requests.post(
        'https://www.jstor.org/api/labs-topic-service/topics',
        headers={
            'User-Agent': 'Python API client',
            'Authorization': 'JWT %s' % access_token},
        json={'text': text}
    ).json()


def decode_token(access_token):
    decoded = jwt.decode(access_token, verify=False)
    # convert expiration timestamp to formatted datetime string
    decoded['exp'] = datetime.fromtimestamp(decoded['exp']).strftime('%Y-%m-%dT%H:%M:%SZ')
    return decoded
```
